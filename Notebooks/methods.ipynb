```python
def read_data(FARSpath):
    # Read all files from FARS
    import pathlib
    from os import walk
    import pandas as pd

    year_dir = []
    for filepath in pathlib.Path(FARSpath).glob('**'):
        year_dir.append(str(filepath.absolute()))

    year_dir.pop(0) # list of all year directories
    
    d = {}
    for year in year_dir:
        print(year)
        csv_files = next(walk(year))[2]
        for file in csv_files:
                dataframe_name = year.split("/")[4][4:8] + "_" + file[:-4]
                csvfile_path = year + "/" + file #2015_accident
                d[dataframe_name] = pd.read_csv(csvfile_path,encoding='ISO8859-1')

    return d
```

```python
def read_accidentdata(FARSpath, d):   
    # Read only 'accident' data from FARS
    import pathlib
    from os import walk
    import pandas as pd
       
    year_dir = []
    for filepath in pathlib.Path(FARSpath).glob('**'):
        year_dir.append(str(filepath.absolute()))

    year_dir.pop(0) # list of all year directories

    for year in year_dir:
        print(year)
        csv_files = next(walk(year))[2]
        for file in csv_files:
            if("accident" in file) or ("ACCIDENT" in file): #read only accident data
                dataframe_name = year.split("/")[4][4:8] + "_" + file[:-4]
                csvfile_path = year + "/" + file #2015_accident
                d[dataframe_name] = pd.read_csv(csvfile_path,encoding='ISO8859-1')
                               
    accident_list = []
    for key in d.keys():
        if ("accident" in key) or ("ACCIDENT" in key):
            accident_list.append(key)
    accident_df_names = []
    for i in accident_list:
        temp_df = d[i]
        accident_df_names.append(temp_df)
    d['accident'] = pd.concat(accident_df_names)
    
    return d['accident']   
```

```python
def read_vehicledata(FARSpath, d):
    # Read only 'vehicle' data from FARS
    import pathlib
    from os import walk
    import pandas as pd
      
    year_dir = []
    for filepath in pathlib.Path(FARSpath).glob('**'):
        year_dir.append(str(filepath.absolute()))

    year_dir.pop(0) # list of all year directories

    for year in year_dir:
        print(year)
        csv_files = next(walk(year))[2]
        for file in csv_files:
            if("vehicle" in file) or ("VEHICLE" in file) or ("Vehicle" in file): #read only vehicle data
                dataframe_name = year.split("/")[4][4:8] + "_" + file[:-4]
                csvfile_path = year + "/" + file
                d[dataframe_name] = pd.read_csv(csvfile_path,encoding='ISO8859-1')
           
    vehicle_list = []
    for key in d.keys():
        if (key.endswith('vehicle')) or (key.endswith('VEHICLE')) or (key.endswith('Vehicle')):
            vehicle_list.append(key)
    vehicle_df_names = []
    for i in vehicle_list:
        temp_df = d[i]
        vehicle_df_names.append(temp_df)

    d['vehicle'] = pd.concat(vehicle_df_names)
    
    return d['vehicle']   
```

```python
def read_persondata(FARSpath, d):   
    # Read only 'person' data from FARS
    import pathlib
    from os import walk
    import pandas as pd
        
    year_dir = []
    for filepath in pathlib.Path(FARSpath).glob('**'):
        year_dir.append(str(filepath.absolute()))

    year_dir.pop(0) # list of all year directories

    for year in year_dir:
        print(year)
        csv_files = next(walk(year))[2]
        for file in csv_files:
            if("person" in file) or ("PERSON" in file) or ("Person" in file): #read only accident data
                dataframe_name = year.split("/")[4][4:8] + "_" + file[:-4]
                csvfile_path = year + "/" + file
                d[dataframe_name] = pd.read_csv(csvfile_path,encoding='ISO8859-1')
                               
    person_list = []
    for key in d.keys():
        if (key.endswith('person')) or (key.endswith('PERSON')) or (key.endswith('Person')):
            person_list.append(key)
    person_df_names = []
    for i in person_list:
        temp_df = d[i]
        person_df_names.append(temp_df)
    d['person'] = pd.concat(person_df_names)
    
    return d['person']   
```

```python
def encode(df, col):
    # to perform one-hot encoding of features
    import pandas as pd
    for i in col:
        df[i] = df[i].astype('category')

    df = pd.get_dummies(df, drop_first=True)
    df = df.fillna(0)   
    return df
```
